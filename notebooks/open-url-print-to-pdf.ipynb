```python
# Import necessary libraries
import pandas as pd
import asyncio
from pyppeteer import launch

# Function to read data from an Excel file
def read_excel(file_path):
    # Load the Excel file
    df = pd.read_excel(file_path)
    return df

# Function to save a webpage as PDF using pyppeteer
def save_as_pdf(url, output_path):
    browser = await launch(headless=True)
    page = await browser.newPage()
    await page.goto(url, {'waitUntil': 'networkidle2'})
    await page.pdf({'path': output_path, 'format': 'A4'})
    await browser.close()

# Main function to process the URLs and save PDFs
def main(excel_path, output_folder):
    # Read the data from Excel
    df = read_excel(excel_path)
    
    # Extract the URL and filename columns
    urls = df['Invoice URL']
    filenames = df['Filename']
    
    # Loop through each URL and save as PDF
    for url, filename in zip(urls, filenames):
        output_path = f"{output_folder}/{filename}.pdf"
        await save_as_pdf(url, output_path)

file_path = 'C:/Users/keith/Dropbox/Embrace/chewy.xlsx'
output_folder = 'C:/Users/keith/Dropbox/Embrace/Downloads'

main(file_path, output_folder)
```

```python
# Import required libraries
import pandas as pd
import pdfkit
import os

# Load the data from an Excel spreadsheet
file_path = 'C:/Users/keith/Dropbox/Embrace/chewy.xlsx'
data = pd.read_excel(file_path)

# Extract URLs and filenames from the data table
urls = data['Invoice URL']  # Replace with your actual column name
filenames = data['Filename']  # Replace with your actual column name

# Define the folder to save PDFs
output_folder = 'C:/Users/keith/Dropbox/Embrace/Downloads'
os.makedirs(output_folder, exist_ok=True)

# Configure PDFKit options
options = {
    'quiet': '',
    "enable-local-file-access": "",
}

# Loop through URLs and save each page as PDF with the specified filename
for url, filename in zip(urls, filenames):
    try:
        output_path = os.path.join(output_folder, f"{filename}.pdf")
        pdfkit.from_url(url, output_path, options=options)
        print(f"Saved {url} as {output_path}")
    except Exception as e:
        print(f"Failed to save {url} as {filename}.pdf: {e}")

print("All URLs processed.")
```

```python
# Read excel file, extract one column of URLs, and download the URLs to a folder.
# Usage: python3 download_urls.py <excel_file> <sheet_name> <column_name> <output_folder>
# Example: python3 download_urls.py data.xlsx Sheet1 URL output

import os
import sys
import pandas as pd
import requests
from urllib.parse import urlparse
from urllib.request import urlretrieve
from tqdm import tqdm

def download_urls(excel_file, sheet_name, column_name, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    df = pd.read_excel(excel_file, sheet_name=sheet_name)
    urls = df[column_name].tolist()

    for url in tqdm(urls):
        try:
            filename = os.path.basename(urlparse(url).path)
            filepath = os.path.join(output_folder, filename)
            urlretrieve(url, filepath)
        except Exception as e:
            print(f"Failed to download {url}: {e}")

if __name__ == "__main__":
    if len(sys.argv) != 4:
        print("Usage: python3 download_urls.py <excel_file> <sheet_name> <column_name> <output_folder>")
        sys.exit(1)

    download_urls(sys.argv[0], sys.argv[1], sys.argv[2], sys.argv[3])

download_urls("C:/Users/keith/Dropbox/Embrace/chewy.xlsx", "2024 To Date", "Invoice URL", "C:/Users/keith/Dropbox/Embrace/Downloads")


```
