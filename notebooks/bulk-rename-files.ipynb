```python
# importing pandas
import pandas as pd

# read text file into pandas DataFrame and create 
df = pd.read_csv("C:/Users/keith/Dropbox/Code/Text_Processing4.txt", header=None)

display(df)

non_url_safe = ['"', '#', '$', '%', '&', '+',
    ',', '/', ':', ';', '=', '?',
    '@', '[', '\\', ']', '^', '`',
    '{', '|', '}', '~', "'"]

safe_names = []

for iloop in range(len(df)):
    txt = df.loc[iloop].to_string(index=False)
#     print(txt)
    txt = txt.replace(' - ', '-')
    txt = txt.replace(' and ', '-')
    txt = txt.replace(' ', '-').lower()
#     print(txt)
    non_safe = [character for character in txt if character in non_url_safe]
    if non_safe:
        for i in non_safe:
            txt = txt.replace(i, '')
    safe_names.append(u'-'.join(txt.split()))

print(safe_names)
```

```python
parent_dir = "C:/Users/keith/Dropbox/Elementsist/Zarlec Website/Products/"
parent_dir = "C:/Users/keith/Downloads/temp/"
for name in safe_names:
    path = os.path.join(parent_dir, name) 
    os.mkdir(path) 
```
