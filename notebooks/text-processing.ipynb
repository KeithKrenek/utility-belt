```python
# Read Text Files with Pandas using read_csv()

# importing pandas
import pandas as pd

# read text file into pandas DataFrame and create 
# header with names
df = pd.read_csv("C:/Users/keith/Dropbox/Code/Text_Processing.txt", sep="/n", header=None)


```

```python
categories = df.iloc[::6]
print(categories)
```

```python
indsapps = pd.DataFrame(df.iloc[1::6]).reset_index(drop=True)

# print(indsapps)

for iloop in range(2,7):
    indsapps = pd.concat([indsapps,pd.DataFrame(df.iloc[iloop::6]).reset_index(drop=True)], axis=1, ignore_index=True)
    
print(indsapps)
```
